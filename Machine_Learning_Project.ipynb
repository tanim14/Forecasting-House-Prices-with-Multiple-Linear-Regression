# 5. Model Selection

As Multiple linear regression allows us to model the relationship between multiple independent variables and a single dependent variable, which is the case in our house_prices dataset because here multiple predictors are influencing the final sale price of the houses. Here, Multiple linear regression can provide insights into which variables are most strongly associated with the outcome and how much each variable contributes to the overall model.

# 6. Preparing the dataset to fit in the model

### 6.1 Scaling the DataFrame

```python
from sklearn.preprocessing import StandardScaler

# create a StandardScaler object
scaler = StandardScaler()

# fit the scaler to the data and transform the data
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data_1 =pd.DataFrame(scaled_data)
```

### 6.2 Defining variables

```python
X= scaled_data_1
y = df['Target']
print("Shape of X = ", X.shape)
print("Shape of y = ", y.shape)
```

### 6.3 Splitting the DataFrame

```python
#importing machine learning library
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2, 
                                                    random_state=50)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# 7. Model training

### 7.1 Model fit

```python
lr = LinearRegression()
model=lr.fit(X_train,y_train)
```

### 7.2 Calulating coefficients

```python
coefficients = model.coef_
print(coefficients)
```

```python
coef_df = pd.DataFrame({"features": X.columns, 
                       "coefficients": np.squeeze(coefficients)})

coef_df
```

### 7.3 Prediction by model

```python
y_pred = model.predict(X_test)
print(y_pred)
```

# 8. Model Evaluation

### 8.1 Error estimates

```python
MAE = mean_absolute_error(y_test, y_pred)
print("MAE = ", MAE)

MSE = mean_squared_error(y_test, y_pred, squared=True)
print("MSE = ", MSE)

RMSE = mean_squared_error(y_test, y_pred, squared=False)
print("RMSE = ", RMSE)

r2 = r2_score(y_test, y_pred)#Goodness of fit-coefficent of determination
print("r_squared = ", r2)
```

### 8.2 Comparison of Actual value and Predicted value

```python
y_pred = model.predict(X_test)

price = pd.DataFrame({"Price_actual":y_test,
                   "Price_predicted": y_pred})

price.head(10)
```
